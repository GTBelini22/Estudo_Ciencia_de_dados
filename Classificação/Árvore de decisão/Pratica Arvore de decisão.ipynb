```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Classificação/Árvore de decisão/golf_df.xlsx')
```

```python
dados.head()
```

```python
x = pd.get_dummies(dados[['clima', 'temp','umidade', 'vento']])
x.head()
```

```python
y = dados[['jogar']]
y
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
modelo = arvore.fit(x,y)
```

```python
modelo
```

```python
# Plotando a árvore
fig = plt.figure(figsize=(50,20))

fig = tree.plot_tree(modelo,
                     feature_names= x.columns,
                     class_names=['yes', 'no'],
                     filled=True
                     )
```
