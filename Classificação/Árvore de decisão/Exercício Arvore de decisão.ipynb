## Fazer a predição se o cliente realizou a compra ou não

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vendas = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Classificação/Árvore de decisão/vendas.xlsx')
```

```python
vendas.head()
```

```python
vendas.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
# Tranformando as variáveis em dummies
le.fit(vendas.test)
vendas.test =  le.transform(vendas.test)
```

```python
vendas.head()
```

```python
x= vendas[['media_rating', 'media_pedidos', 'test']]

y = vendas[['cliente_comprou']]
```

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
modelo = arvore.fit(x,y)
```

```python
modelo
```

```python
fig = plt.figure(figsize=(15,10))

fig =  tree.plot_tree(decision_tree=modelo, feature_names=x.columns, class_names=['yes', 'no'], filled=True)
```
