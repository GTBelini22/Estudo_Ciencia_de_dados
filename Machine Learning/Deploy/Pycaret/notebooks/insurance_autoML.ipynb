```python
import os 
import pandas as pd
from pycaret.regression import *
```

```python
FEATURES = ['age', 'bmi', 'children', 'smoker']
TARGET = 'charges'
```

```python
data =  pd.read_csv('../data\insurance.csv')
data =  data[[*FEATURES, TARGET]]
data
```

```python
# session_id é igual ao random_state
s = setup(data, target='charges', session_id=2024)
```

```python
# Comparar varios modelos
best_model = compare_models()
```

```python
best_model
```

```python
tune_model =  tune_model(best_model, n_iter=100)
```

```python
evaluate_model(best_model)
```

```python
final_model = finalize_model(best_model)
```
