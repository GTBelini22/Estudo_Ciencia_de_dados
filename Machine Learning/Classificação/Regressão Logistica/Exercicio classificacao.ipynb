## Predizendo do dataset de golf a coluna de jogar

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
golf = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Classificação/Regressão Logistica/golf_df.xlsx')
```

```python
golf.head(10)
```

```python
golf.info()
```

Tranformando as colunas categoricas em numericas

```python
le = LabelEncoder()
```

```python
golf.columns
```

```python
le.fit(golf.clima)
golf.clima = le.transform(golf.clima)

le.fit(golf.temp)
golf.temp = le.transform(golf.temp)

le.fit(golf.umidade)
golf.umidade = le.transform(golf.umidade)

le.fit(golf.vento)
golf.vento = le.transform(golf.vento)
```

```python
golf.head()
```

Separando as bases

```python
x = golf.iloc[:, :-1]

y= golf.iloc[:, -1]
# OU
# y = golf[['jogar']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
golf_pred = model.predict(x_test)
```

```python
x_test['Predicao'] = golf_pred

y_test = pd.DataFrame(y_test)


x_test['real'] = y_test['jogar']
x_test.head(20)
```
