```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Regressão Linear/Otimizção/house.xlsx')
```

```python
df.head()
```

```python
df.info()
# Visto a existencia de nulos
```

```python
df = df.dropna()
```

```python
# Colocando em variáveis dummies
df = pd.get_dummies(df)
df.head()
```

```python
df.columns
```

```python
# ajustando para usar com o Gradient boost

labels = np.array(df['SalePrice'])

df = df.drop('SalePrice', axis=1)

df_list = list(df.columns)

df = np.array(df)
```

```python
# Treinando o modelo

from sklearn.model_selection import train_test_split

train_df, test_df, train_labels, test_labels = train_test_split(df, labels, test_size=0.25, random_state=42)
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

##GradientBoostRegressor

```python
gbr = GradientBoostingRegressor(n_estimators=100)

gbr.fit(train_df, train_labels)
gbr_pred = gbr.predict(test_df)
```

```python
errors_gbr = abs(gbr_pred - test_labels)

r2 = gbr.score(test_df, test_labels)
print(r2)

print('MAE: ', metrics.mean_absolute_error(test_labels, gbr_pred))
```

##Iniciando os hiperparametros

```python
from sklearn.model_selection import GridSearchCV
```

```python

parameters = {'learning_rate': [0.03],
                  'subsample'    : [0.2],
                  'n_estimators' : [500,1000, 1500],
                  'max_depth'    : [8]
                 }

grid_search = GridSearchCV(gbr, parameters, scoring='r2', cv=2, n_jobs=-1)
```

```python
grid_search.fit(train_df, train_labels)
```

```python
# Plotando os melhores parâmetros
print('melhor estimador:', grid_search.best_estimator_)
print('melhor score:', grid_search.best_score_)
print('melhor parametro:', grid_search.best_params_)
```
