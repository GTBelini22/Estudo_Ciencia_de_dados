```python
# Arrumando as colunas obsjects com Dummies

df = pd.get_dummies(df)
df.head()
```

Separando as colunas de X e Y para a analise

```python
df.columns
```

```python
sale = np.array(df['SalePrice'])

feature = df.drop('SalePrice', axis=1)
# Atribuindo a lista de colunas
feature_list = list(df.columns)

df = np.array(df)
```

```python
from sklearn.model_selection import train_test_split

train_feature, test_feature, train_sale, test_sale = train_test_split(feature,sale, test_size=0.25, random_state=42)
```

Treinando o modelo

```python
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

##Random Forest Reegressor

```python
# Configuração do Forest regressor
rf = RandomForestRegressor(n_estimators=1000, random_state=42)
rf.fit(train_feature, train_sale)
```

```python
# Modelo treinado, agora fazendo a predição
prediction_rf = rf.predict(test_feature)

r_sq = rf.score(feature, sale)

print('R2: ', r_sq)
print('MAE: ', metrics.mean_absolute_error(test_sale, prediction_rf))
print('MSE: ', metrics.mean_squared_error(test_sale, prediction_rf))
```

## Ada Boost Reegressor

```python
# Configurando o Ada
ada = AdaBoostRegressor(n_estimators=100)

ada.fit(train_feature, train_sale)
```

```python
# Criando as predições
prediction_ada = ada.predict(test_feature)

r_sq = ada.score(feature,sale)

print('R2: ', r_sq)
print('MAE: ', metrics.mean_absolute_error(test_sale, prediction_ada))
print('MSE: ', metrics.mean_squared_error(test_sale, prediction_ada))
```

## Gradient boost

```python
# Configurando o gradient boost
gbr = GradientBoostingRegressor(n_estimators=100)

gbr.fit(train_feature, train_sale)
```

```python
prediction_gbr = gbr.predict(test_feature)

r_sq = gbr.score(feature,sale)

print('R2: ', r_sq)
print('MAE: ', metrics.mean_absolute_error(test_sale, prediction_gbr))
print('MSE: ', metrics.mean_squared_error(test_sale, prediction_gbr))

# Gradiente boost performou melhor que os demais
```
